```python
import turtle

wn = turtle.Screen()
wn.title("Pong by @MichelleRuas")
wn.bgcolor("blue")
wn.setup(width=800, height=600)
wn.tracer(0)


#Beaver    
beaver = turtle.Turtle()
beaver.clear()
beaver.speed(0)
beaver.color("white")
beaver.shape("circle")
beaver.penup()
beaver.goto(-350, 250)


#Obstacle 1   
obstacle1 = turtle.Turtle()
obstacle1.clear()
obstacle1.speed(0)
obstacle1.color("black")
obstacle1.shape("square")
obstacle1.shapesize(stretch_wid = 1, stretch_len = 8)
obstacle1.penup()
obstacle1.goto(200, 100)
obstacle1.dx = 3

#Obstacle 2   
obstacle2 = turtle.Turtle()
obstacle2.clear()
obstacle2.speed(0)
obstacle2.color("black")
obstacle2.shape("square")
obstacle2.shapesize(stretch_wid = 1, stretch_len = 10)
obstacle2.penup()
obstacle2.goto(-100,0)
obstacle2.dx = 4

#Obstacle 3   
obstacle3 = turtle.Turtle()
obstacle3.clear()
obstacle3.speed(0)
obstacle3.color("black")
obstacle3.shape("square")
obstacle3.shapesize(stretch_wid = 1, stretch_len = 4)
obstacle3.penup()
obstacle3.goto(-200,-100)
obstacle3.dx = 5

#Obstacle 4 
obstacle4 = turtle.Turtle()
obstacle4.clear()
obstacle4.speed(0)
obstacle4.color("black")
obstacle4.shape("square")
obstacle4.shapesize(stretch_wid = 1, stretch_len = 12)
obstacle4.penup()
obstacle4.goto(50,-200)
obstacle4.dx = 3

#Home
home = turtle.Turtle()
home.speed(0)
home.color("yellow")
home.shape("circle")
home.shapesize(stretch_wid = 2, stretch_len = 2)
home.penup()
home.goto(300,-250)


#Pen
pen = turtle.Turtle()
pen.speed(0)
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0,0)
pen.clear()


#Movement functions
def beaver_up():
    y = beaver.ycor()
    y += 20
    beaver.sety(y)
    
def beaver_down():
    y = beaver.ycor()
    y -= 20
    beaver.sety(y)

def beaver_right():
    x = beaver.xcor()
    x += 20
    beaver.setx(x)
    
def beaver_left():
    x = beaver.xcor()
    x -= 20
    beaver.setx(x)
    
    
#Keyboard binding
wn.listen()
wn.onkeypress(beaver_up, "Up")
wn.onkeypress(beaver_down, "Down")
wn.onkeypress(beaver_left, "Left")
wn.onkeypress(beaver_right, "Right")

while True:
    wn.update()
    obstacle1.setx(obstacle1.xcor() + obstacle1.dx)
    obstacle2.setx(obstacle2.xcor() + obstacle2.dx)
    obstacle3.setx(obstacle3.xcor() + obstacle3.dx)
    obstacle4.setx(obstacle4.xcor() + obstacle4.dx)
    
    if obstacle1.xcor() > 300:
        obstacle1.goto(300, 100)
        obstacle1.dx *= -1
        
    if obstacle1.xcor() < -300:
        obstacle1.goto(-300, 100)
        obstacle1.dx *= -1
        
    if (beaver.ycor() > 80 and beaver.ycor() < 120) and (beaver.xcor() < obstacle1.xcor() + 80 and beaver.xcor() > obstacle1.xcor() - 80):
        beaver.goto(-350, 250)
        
    if obstacle2.xcor() > 300:
        obstacle2.goto(300, 0)
        obstacle2.dx *= -1
        
    if obstacle2.xcor() < -300:
        obstacle2.goto(-300, 0)
        obstacle2.dx *= -1
        
    if (beaver.ycor() > -20 and beaver.ycor() < 20) and (beaver.xcor() < obstacle2.xcor() + 90 and beaver.xcor() > obstacle2.xcor() - 90):
        beaver.goto(-350, 250)    
        
    if obstacle3.xcor() > 300:
        obstacle3.goto(300, -100)
        obstacle3.dx *= -1
        
    if obstacle3.xcor() < -300:
        obstacle3.goto(-300, -100)
        obstacle3.dx *= -1
        
    if (beaver.ycor() > -120 and beaver.ycor() < -80) and (beaver.xcor() < obstacle3.xcor() + 60 and beaver.xcor() > obstacle3.xcor() - 60):
        beaver.goto(-350, 250)
        
    if obstacle4.xcor() > 300:
        obstacle4.goto(300, -200)
        obstacle4.dx *= -1
        
    if obstacle4.xcor() < -300:
        obstacle4.goto(-300, -200)
        obstacle4.dx *= -1
        
    if (beaver.ycor() > -220 and beaver.ycor() < -180) and (beaver.xcor() < obstacle4.xcor() + 130 and beaver.xcor() > obstacle4.xcor() - 130):
        beaver.goto(-350, 250)
        
    if (beaver.ycor() < home.ycor() + 20 and beaver.ycor() > home.ycor() - 20) and (beaver.xcor() < home.xcor() + 20 and beaver.xcor() > home.xcor() - 20):
        pen.write("You Win!!! Congrats :)", font = ("Courier", 50, "normal"), align = "center")


   



```
